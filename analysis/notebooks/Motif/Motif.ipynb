```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import cm
import seaborn as sns
import sim_utils as s_utils
import analysis_utils as a_utils
import h5py
from scipy import signal
```

# Autonomous Osc

```python
sim_id = 'MotifAutoOsc'
params=s_utils.load_sim_params(sim_id)
stell_spikes_l,intrnrn_spikes_l=s_utils.load_spikes(sim_id=sim_id)
stell_v = s_utils.load_full_data(sim_id,data_id="stell_v")
intrnrn_v = s_utils.load_full_data(sim_id,data_id="intrnrn_v")
t = np.linspace(0,params["sim_dur"],len(intrnrn_v[0]))
```

```python
%matplotlib inline
plt.style.use("analysis/config/paper.mplstyle")
fig,axs = plt.subplots(2,1,figsize=(8,4))
axs[0].plot(t,stell_v[0])
axs[0].plot(t,stell_v[1])
axs[1].plot(t,intrnrn_v[0])
axs[1].plot(t,intrnrn_v[1])
axs[1].set_ylim([-90,80])
axs[0].set_ylim([-90,80])

plt.show()
```

# Switch

```python
sim_id = 'MotifSwitch'
params=s_utils.load_sim_params(sim_id)
stell_spikes_l,intrnrn_spikes_l=s_utils.load_spikes(sim_id=sim_id)
stell_v = s_utils.load_full_data(sim_id,data_id="stell_v")
intrnrn_v = s_utils.load_full_data(sim_id,data_id="intrnrn_v")
input_amp = s_utils.load_full_data(sim_id,data_id="intrnrn_switch_pulse")

t = np.linspace(0,params["sim_dur"],len(intrnrn_v[0]))
```

```python
%matplotlib inline
plt.style.use("analysis/config/paper.mplstyle")
fig,axs = plt.subplots(3,1,figsize=(12,4))

axs[0].plot(t,stell_v[0],label="stell_0")
axs[0].plot(t,stell_v[1],label="stell_1")
axs[1].plot(t,intrnrn_v[0],label="intrnrn_0")
axs[1].plot(t,intrnrn_v[1],label="intrnrn_1")


axs[2].plot(t,input_amp[0])
axs[2].plot(t,input_amp[1])

axs[1].set_ylim([-90,80])
axs[0].set_ylim([-90,80])
axs[0].legend()
axs[1].legend()
plt.show()
```
